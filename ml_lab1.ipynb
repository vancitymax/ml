```python
import csv
import numpy as np
import matplotlib.pyplot as plt


trainFile = "lab_1_train.csv"
testFile = "lab_1_test.csv"

def readCSVFile(fileName):
 
    with open(fileName, 'r') as file:
        reader = csv.reader(file)
        next(reader) 
        data = [row for row in reader]
    return data
```

```python
trainDataset = readCSVFile ( trainFile)
```

```python
trainDataset
```

```python
feature = [float(row[1]) for row in trainDataset]
feature
```

```python
target = [float(row[2]) for row in trainDataset]
target
```

```python
plt.title("Train dataset")
plt.plot(feature, target, "o")
```

```python
def linearRegression(b0, b1, x):
   
    return b0 + b1 * x

def mseLossFunction(b0, b1, features, targets):
   
    predictedTargets = [linearRegression(b0, b1, x) for x in features]
    mseLoss = np.mean([(expected - predicted) ** 2 for (expected, predicted) in zip(targets, predictedTargets)])
    return mseLoss
```

```python
tolerance = 0.000001
lr = 0.005

np.random.seed(42)


b0 = float(abs(np.random.randn(1)))
b1 = float(abs(np.random.randn(1)))
loss = mseLossFunction(b0, b1, feature, target)


dt = loss
i = 0


while dt > tolerance:
    predictedTargets = [linearRegression(b0, b1, f) for f in feature]
    predictedActualDiff = [(actual - predicted) for (actual, predicted) in zip(target, predictedTargets)]
    dB0 = -2  * np.sum(predictedActualDiff) / len(feature)
    dB1 = (-2 / len(feature)) * np.sum([(x * diffValue) for (x, diffValue) in zip(feature, predictedActualDiff)])
    b0 = b0 - lr * dB0
    b1 = b1 - lr * dB1
    print(f'{i}: b0 = {b0}; b1 = {b1}; loss = {loss};')
    i += 1
    newLoss = mseLossFunction(b0, b1, feature, target)
    dt = abs(loss - newLoss)
    loss = newLoss
```

```python
print("--------------------Results------------------")
print("Numbers of iterations = ", i)
print("B0 = ", b0)
print("B1 = ", b1)
print("Loss function= ", loss)
```
